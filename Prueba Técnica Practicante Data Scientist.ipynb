**Clustering K-means**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Convertir los títulos de las categorías en valores numéricos
le = LabelEncoder()
df_cleaned['title_encoded'] = le.fit_transform(df_cleaned['items__snippet__title'])

# Realizar clustering con los ID
X = df_cleaned[['items__id', 'title_encoded']]

# Aplicar k-means clustering con 3 grupos
kmeans = KMeans(n_clusters=3, random_state=42)
df_cleaned['cluster'] = kmeans.fit_predict(X)

# Mostrar las categorías con sus respectivos clusters
print(df_cleaned[['items__id', 'items__snippet__title', 'cluster']])

# Visualización del clustering
plt.scatter(df_cleaned['items__id'], df_cleaned['title_encoded'], c=df_cleaned['cluster'], cmap='viridis')
plt.xlabel('ID de Categoría')
plt.ylabel('Cantidad')
plt.title('Clustering de Categorías')
plt.show()
```

**Modelo de Clasificación**

**Random forest**

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import numpy as np

# Crear el target (variable a predecir): "assignable"
y = df_cleaned['items__snippet__assignable'].astype(int)

# Usar los IDs de las categorías y los títulos codificados como características
df_cleaned['category_encoded'] = df_cleaned['items__snippet__title'].astype('category').cat.codes
X = df_cleaned[['items__id', 'category_encoded']]

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de Árbol de Decisión
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Realizar predicciones
y_pred = clf.predict(X_test)

# Reporte de clasificación
print("Reporte de Clasificación:")
print(classification_report(y_test, y_pred))

# Visualización del Árbol de Decisión
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
tree.plot_tree(clf, feature_names=['items__id', 'category_encoded'], class_names=['No Asignable', 'Asignable'], filled=True)
plt.title("Árbol de Decisión para Clasificación de Videos")
plt.show()
```

**Modelo Probabilístico**

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix

# Crear el modelo Naive Bayes
gnb = GaussianNB()

# Entrenar el modelo con los mismos datos que usamos para el Árbol de Decisión
gnb.fit(X_train, y_train)

# Realizar predicciones
y_pred_nb = gnb.predict(X_test)

# Evaluación del modelo
print("Precisión de Naive Bayes:", accuracy_score(y_test, y_pred_nb))
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_nb))

# Reporte de Clasificación
print("Reporte de Clasificación (Naive Bayes):")
print(classification_report(y_test, y_pred_nb))
```

**Forecasting arima**

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA  # Importar la nueva implementación de ARIMA
import matplotlib.pyplot as plt
import matplotlib.dates as mdates  # Importar para formatear las fechas

# Simulación de datos de vistas y fechas (si tuviéramos fechas y vistas reales)
df_cleaned['views'] = np.random.randint(1000, 100000, df_cleaned.shape[0])  # Simulación de vistas
df_cleaned['publish_date'] = pd.date_range(start='2023-01-01', periods=df_cleaned.shape[0], freq='W')  # Simulación de fechas

# Ordenar los datos por fecha
df_cleaned = df_cleaned.sort_values('publish_date')

# Crear una serie temporal usando las vistas
series = df_cleaned.set_index('publish_date')['views']

# Ajustar el modelo ARIMA
model = ARIMA(series, order=(1, 1, 1))
model_fit = model.fit()

# Resumen del modelo
print(model_fit.summary())

# Predecir las próximas 10 semanas
forecast = model_fit.forecast(steps=10)
print("Pronóstico de las próximas vistas:", forecast)

# Visualización de la serie temporal y el pronóstico
plt.figure(figsize=(10, 5))  # Ajustar tamaño de la figura
plt.plot(series, label="Datos reales")
plt.plot(pd.date_range(start=series.index[-1], periods=10, freq='W'), forecast, color='red', label="Pronóstico")
plt.legend()
plt.title("Forecasting de vistas en videos")

# Formatear las fechas en el eje x
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Muestra un tick por mes
plt.gcf().autofmt_xdate()  # Ajusta el formato de la fecha para evitar superposiciones

plt.show()
```

**Regresion lineal**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

# Simulación de datos de likes y vistas
df_cleaned['likes'] = np.random.randint(100, 10000, df_cleaned.shape[0])  # Simulación de likes
df_cleaned['views'] = np.random.randint(1000, 1000000, df_cleaned.shape[0])  # Simulación de vistas

# Codificar los títulos de las categorías
df_cleaned['category_encoded'] = df_cleaned['items__snippet__title'].astype('category').cat.codes

# Selección de características: el ID de la categoría y si es asignable
X = df_cleaned[['items__id', 'category_encoded']]

# Variable a predecir like
y = df_cleaned['likes']

# División del dataset en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Predicción de likes
y_pred = model.predict(X_test)

# Evaluación del modelo R²
print(f"Coeficiente de Determinación (R²): {model.score(X_test, y_test)}")

# Visualización de los resultados de predicción vs valores reales
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.xlabel("Likes reales")
plt.ylabel("Likes predichos")
plt.title("Predicción de Likes usando Regresión Lineal")
plt.show()
```
